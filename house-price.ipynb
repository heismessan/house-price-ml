```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
```

LECTURE DES FICHIERS

```python
data = pd.read_csv('drive/MyDrive/archive/data.csv')
dt = data.copy()
dt = dt.drop(["country"], axis=1)
#extraction du mois
dt["date"] = pd.to_datetime(dt["date"], errors='coerce')
dt["month"] = dt["date"].dt.month
dt.head()
```

```python

train = dt[dt["month"] < 7]
test = dt[dt["month"] == 7]
train["age_house"] = 2025 - train["yr_built"]
test["age_house"] = 2025 - test["yr_built"]
train["recently_renovation"] = train["yr_renovated"] > 0
test["recently_renovation"] = test["yr_renovated"] > 0




X_train = train.drop(["price", "date"], axis=1)
y_train = train["price"]
X_test = test.drop(["price", "date"], axis=1)
y_test = test["price"]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_scale = sc.fit_transform(X_train)
X_test_scale = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train_scale, y_train).score(X_train_scale, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
y_pred = reg.predict(X_test_scale)
MAE = mean_absolute_error(y_test, y_pred)
MSE = mean_squared_error(y_test, y_pred)
RMSE = np.sqrt(MSE)
r2 = r2_score(y_test, y_pred)
print("MAE: ", MAE)
print("MSE: ", MSE)
print("RMSE: ", RMSE)
print("R2: ", r2)
```

FEATURE ENGENEERING

```python
num_feat = [
    'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
       'floors', 'waterfront', 'view', 'condition', 'sqft_above',
       'sqft_basement', 'yr_built', 'yr_renovated', 'month'
]
cat_feat = [
    'city', 'statezip'
]
```

from sklearn.tree import DecisionTreeRegressor

from sklearn.ensemble import RandomForestRegressor


```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor

models = {
    "DecisionTreeRegressor": DecisionTreeRegressor(),
    "RandomForestRegressor": RandomForestRegressor(),
    "GradientBoostingRegressor": GradientBoostingRegressor(),
    "XGBRegressor": XGBRegressor()
}
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_feat),
        ('cat', OneHotEncoder(), cat_feat)
    ]
)
results = {}
for name, model in models.items():
  pipe = Pipeline(
      steps=[
          ('preprocessor', preprocessor),
          ('regression', model)
      ]
  )

  pipe.fit(X_train, y_train)
  pipe.score(X_train, y_train)
  y_pred = pipe.predict(X_test)
  r2 = r2_score(y_test, y_pred)
  results[name] = r2
```

```python
for model, r2 in results.items():
  print(f"{model} : {r2:.4f}")
```
